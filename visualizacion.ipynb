```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
custom_date_parser = lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S")
url='https://github.com/SebastianOrtega/datasets/blob/main/weight.csv?raw=True'
df=pd.read_csv(url, parse_dates=['Date'],date_parser=custom_date_parser)
df=df.reindex(index=df.index[::-1])
df.drop(index=1047, inplace=True)
df.head()
```

```python
#x = np.linspace(0, 2, 100)
# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
start_date = '2019-3-31'
x=df[df.Date < start_date].Date
#x=df.index
y=df[df.Date < start_date]['Peso (kg)']

fig, ax = plt.subplots(figsize=(20, 10.7))
ax.plot(x, y, label='peso')  # Plot some data on the axes.
#ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
#ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

```python
fig, axes = plt.subplots(3, 2)
sns.set(rc={'figure.figsize':(20,20)})
sns.boxplot(x=ndf.Date.dt.month, width=.8, y="Peso (kg)", data=ndf , ax=axes[0,0])
sns.boxplot(x=ndf.Date.dt.month, width=.8, y="Masa grasa (kg)", data=ndf, ax=axes[0,1])
sns.boxplot(x=ndf.Date.dt.month, width=.8, y="Muscle mass (kg)", data=ndf, ax=axes[1,0])
sns.boxplot(x=ndf.Date.dt.month, width=.8, y="Bone mass (kg)", data=ndf, ax=axes[1,1])
sns.boxplot(x=ndf.Date.dt.month, width=.8, y="Hidratación (kg)", data=ndf, ax=axes[2,0])
plt.show()
```
